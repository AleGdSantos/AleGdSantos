### Legenda dos Dados

**PassengerId:** Este é um identificador único para cada passageiro. Ele não contém informações específicas sobre o passageiro, mas é útil para rastrear registros individuais.

**Fare (Tarifa):** Representa o valor pago pelo passageiro pela passagem.

**Cabin (Cabine):** Indica a cabine em que o passageiro estava alojado, quando disponível. Muitos registros estão ausentes.

**Embarked (Porto de Embarque):** Mostra o local de embarque do passageiro, com valores "C" para Cherbourg, "Q" para Queenstown e "S" para Southampton.

#### Tratando Valores Vazios (NAN)

Tratar valores nulos é fundamental para garantir a qualidade e precisão das análises de dados. Valores nulos podem distorcer resultados e afetar a interpretação. As principais alternativas para tratar valores nulos são:

**Remoção de Linhas:** Eliminar as linhas que contenham valores nulos, mas isso pode levar à perda de dados.

**Preenchimento:** Preencher os valores nulos com dados apropriados, como a média, mediana ou um valor específico, com base no contexto.

**Interpolação:** Estimar valores nulos com base em valores adjacentes, útil para dados sequenciais.

**Modelagem Preditiva:** Usar algoritmos de aprendizado de máquina para prever valores nulos com base em outros atributos.

A escolha depende do contexto e do impacto nos resultados da análise.

**Variáveis Dummy**

Para as colunas Pclass e Embarked aplicaremos a função Dummy.
Essas variáveis permitem a representação e análise apropriada de categorias, garantindo que informações relevantes sejam preservadas e usadas eficazmente em diversas aplicações analíticas, enquanto atendem aos requisitos de algoritmos que exigem dados numéricos.

```
# Aqui separamos as colunas previsoras, que no caso será Survived
previsores = df3[['Family', 'Sex', 'Pclass',
       'Age', 'Fare', 'Embarked']]
```

```
# Aqui separamos a coluna alvo, que no caso será Survived
alvo = df3['Survived']
alvo
```

Nesta etapa faremos uma separação dos atributos para facilitar a construção das variáveis dummy na coluna ingredients.

```
# Aplicando get_dummies nas colunas 'Embarked' e 'Pclass'
dummies_embarked = pd.get_dummies(previsores['Embarked'], prefix='Embarked')
dummies_pclass = pd.get_dummies(previsores['Pclass'], prefix='Pclass')

# Concatenando as colunas dummy ao DataFrame original
df3 = pd.concat([previsores, dummies_embarked, dummies_pclass], axis=1)

# Exibindo as três primeiras linhas do DataFrame resultante
df3.head(3)

```
