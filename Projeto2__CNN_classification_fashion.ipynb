## Importar Bibliotecas

```
#Numpy: operações matemáticas
import numpy as np
#Matplotlib e Seaborn: Plotagem de gráficos
import matplotlib.pyplot as plt
import seaborn as sns
#Sklearn: funções de aprendizagem de máquina
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
#Tensor FLow e Keras: Construção da rede neural
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPool2D, Flatten, Dense, Dropout
```

```
fashion_mnist = keras.datasets.fashion_mnist

(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

##  Pré-processamento dos dados

###Normalização

```
train_images = train_images / 255.0

test_images = test_images / 255.0
```

### Remodelagem

```
train_images = train_images.reshape(60000,28,28,1)

test_images = test_images.reshape(10000,28,28,1)
```

## **Rede Neural Convolucional (CNN)**

**Rede Neural**

Uma rede neural é um modelo computacional inspirado no sistema nervoso biológico, capaz de aprender e realizar tarefas através de um conjunto de conexões entre unidades de processamento chamadas de neurônios artificiais.

camadas_an.gif

**Rede Neural Convolucional**

Uma Convolutional Neural Network (CNN) é um tipo de rede neural especializada em processamento de imagens, usando camadas convolucionais para extrair características e classificar padrões nas imagens.

cat.gif

### Arquitetura de uma Rede Neural Convolucional

**Rede Neural Convolucional (Convolutional Neural Network – CNN)**  comumente aplicada na classificação de imagem, se caracteriza por conseguir distinguir as principais características em uma imagem, gerando uma classificação capaz de generalizar independente da formatação, rotação e dimensões.

  **Operador Convolucional:** É o processo de adicionar cada elemento da imagem
para seus vizinhos, ponderado por um kernel/filtro. Em outras palavras é a combinação de duas fontes de informação, na tentativa de identificar padrões em imagens.


WhatsApp Image 2023-06-07 at 07.59.29.jpeg

**Camada de Convolução**: É o conjunto de mapas de características, sendo a própria rede responsável por decidir qual detector de características será melhor para o modelo.

conv.png

**Pooling**: Aplicado para enfatizar as características da imagem a ser classificada, sendo aplicado para reduzir a sensibilidade dos mapas de características, que poderão ser afetados por fatores externos da imagem.

Nesse filtro iremos selecionar os pontos com maior importância, reduzindo a quantidade de ruídos e possíveis overfits.





pool.png

**Dropout**: função para a regularização de redes neurais, onde alguns neurônios são removidos com a finalidade de evitar um overfitting.

Schematic-diagram-of-Dropout-a-a-standard-neural-network-b-the-neural-network-using.png

**Flattening**: função de achatamento, é aplicada para transformar a matriz de uma imagem em um vetor. Tal vetor será aplicado a camâda de entrada da rede neural.

flat.png

### **Construindo o Modelo**

**Neurônios na camada de entrada**

```
# Neurônios na camada de entrada, será igual a quantidade de pixels em uma imagem
# No caso 28x28
28*28
```

**Neurônios camada oculta**

O número de neurônios na camada oculta normalmente é definido de forma empírica, algumas fontes recomendam ao menos adotar um valor menor que o dobro do valor utilizado na camada de entrada, assim podemos evitar problemas como os de overffiting.

Autores como **Jeff Heaton**, de ***Introduction to Neural Networks for Java***, sugere as seguintes abordagens:

```
# 1ª nº de neurônios na entrada + número de classes /  número de classes
(784+10)/10
```

```
# 2ª 2/3 do tamanho da camada de entrada
(784*2)/(3+10)
```

```
model = keras.Sequential([
    #Camada de convolução
    keras.layers.Conv2D(filters=64, kernel_size= 3, activation='relu'),
    keras.layers.MaxPool2D(pool_size= 2),
    keras.layers.Dropout(0.2),

    keras.layers.Conv2D(filters=32, kernel_size= 3, activation='relu'),
    keras.layers.MaxPool2D(pool_size= 2),
    keras.layers.Dropout(0.2),

    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(10, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])
```

```
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```
history = model.fit(train_images, train_labels, epochs=10)
```

```
plt.plot(history.history['accuracy']);

plt.legend(labels=['modelo1'], title = "Acurácia Treinamento");
```

## Avaliação da rede neural

```
previsores = model.predict(test_images)
```

```
previsores = np.argmax(previsores, axis = 1)
```

```
accuracy_score(test_labels, previsores)
```

```
fig, ax = plt.subplots(figsize=(10, 10))
cm = confusion_matrix(test_labels, previsores)
sns.heatmap(cm, annot=True, fmt='g',cmap="BuPu", ax=ax);
# labels, title and ticks
ax.set_xlabel('Predicted labels');ax.set_ylabel('True labels');
ax.set_title('Confusion Matrix');
ax.xaxis.set_ticklabels(['Camiseta', 'Calças', 'Pulôver', 'Vestido', 'Casaco',
               'Sandália', 'Camisa', 'Tênis', 'Bolsa', 'Bota'],rotation=45);
ax.yaxis.set_ticklabels(['Camiseta', 'Calças', 'Pulôver', 'Vestido', 'Casaco',
               'Sandália', 'Camisa', 'Tênis', 'Bolsa', 'Bota'],rotation=360);
```

```
from sklearn.metrics import classification_report
print(classification_report(test_labels, previsores,target_names=class_names))
```

**Precision :** A precisão é uma medida de quantas das previsões positivas feitas estão corretas (verdadeiros positivos)



**Recall/Sensibilidade:** é uma medida de quantos casos positivos o classificador previu corretamente, em relação a todos os casos positivos no conjunto de dados.



**F1-Score:** é a junção da precisão com a sensibilidade.

## Modelos Adicionais

```
model2 = keras.Sequential([
    keras.layers.Conv2D(filters=32, kernel_size= 2, activation='leaky_relu'),
    keras.layers.MaxPool2D(pool_size= (2,2)),

    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(128, activation='leaky_relu'),
    keras.layers.Dense(10, activation='softmax')
])
```

```
from keras.optimizers import RMSprop
model2.compile(optimizer=RMSprop(learning_rate = 0.001),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```
history2 = model2.fit(train_images, train_labels, epochs=10)
```

```
model3 = keras.Sequential([
    #Camada de convolução
    keras.layers.Conv2D(filters=64, kernel_size= (3,3), activation='leaky_relu'),
    keras.layers.MaxPool2D(pool_size= 2),
    keras.layers.Dropout(0.2),

    #Flatten: convertendo de uma matriz para um vetor - cada um dos valores presentes no vetor serão aplicados na entrada de uma rede neural densa
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(79, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])
```

```
model3.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy']) # compila a rede neural
```

```
history3 = model3.fit(train_images, train_labels, epochs=10)
```
